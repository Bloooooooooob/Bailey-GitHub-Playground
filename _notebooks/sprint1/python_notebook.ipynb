```python
Python Notebook

Basic Python Code

print("")
Displays whatever is in the parentheses in the terminal

variableName = content
Creates a variable and assigns content to it

str(number)
Converts a number to a string, useful for combining with text or printing

int(stringOrSomething)
Converts whatever is in the parentheses into an integer

float(stringOrSomething)
Converts whatever is in the parentheses into a floating-point number

def funcName(var1, var2, etc...):
Defines a function that can take arguments and execute code

funcName(var1, var2, etc...)
Calls the function and runs the code inside of it

return value
Returns a value from a function

Input and Output

input("Prompt")
Gets user input from the console (always returns a string)

len(collection)
Returns the length of a collection (like a string, list, etc.)

Control Flow (If Statements)

if term:
If term is true, runs the code inside the block

elif term:
Checks another condition if the previous if was false

else:
Runs this code if none of the previous if or elif conditions are true

if term1 == term2:
If term1 is equal to term2, runs the code

if term1 != term2:
If term1 is not equal to term2, runs the code

if term1 < term2:
If term1 is less than term2, runs the code

if term1 <= term2:
If term1 is less than or equal to term2, runs the code

if term1 > term2:
If term1 is greater than term2, runs the code

if term1 >= term2:
If term1 is greater than or equal to term2, runs the code

if term1 and term2:
If both term1 and term2 are true, runs the code

if term1 or term2:
If either term1 or term2 is true, runs the code

Loops

while condition:
Keeps running the code inside as long as condition is true

for item in collection:
Iterates through each item in a collection (like a list, string, or range)

break
Stops the loop immediately

continue
Skips the rest of the current loop iteration and moves to the next one

Lists

myList = []
Creates an empty list

myList.append(value)
Adds a value to the end of the list

myList[index]
Accesses the value at a specific index in the list

myList[index] = newValue
Updates the value at a specific index in the list

del myList[index]
Deletes the value at a specific index in the list

Importing Libraries

import moduleName
Imports an external library/module into your program

from moduleName import something
Imports a specific function or object from a module
```
